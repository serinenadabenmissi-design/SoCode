```python
#part 1
import pandas as pd

df=pd.read_csv("iris.csv")

print(df.describe())
```

```python
#part 2
import seaborn as sns
import matplotlib.pyplot as plt


df.dropna().hist(figsize=(10, 8))
plt.show()


sns.pairplot(df.dropna(), hue='target')
plt.show()
```

```python
sns.boxplot(data=df.dropna())
plt.show()
```

```python
#part3
df_no_dplc = df.drop_duplicates()
print(df_no_dplc.describe())
```

```python
print(f"empty cells: {df.isnull().sum()}")

df_cleaned = df.dropna()
print(df_cleaned.head())

df_filled = df.fillna(df.mean())
print(df_filled.head())
```

```python
#part 4

from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler_min_max = MinMaxScaler()
scaler_standard = StandardScaler()

data_normalized = scaler_min_max.fit_transform(df_filled.drop(columns=['target']))
data_standardized = scaler_standard.fit_transform(df_filled.drop(columns=['target']))

df_normalized = pd.DataFrame(data_normalized, columns=['sepal length (cm)',
                        'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
df_standardized = pd.DataFrame(data_standardized, columns=['sepal length (cm)',
                        'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
print(df_normalized)
print(df_standardized)

df_normalized.to_csv("normalized_iris.csv", index=False)
df_standardized.to_csv("standardized_iris.csv", index=False)
```
